Now we want to plot the amount of funding for each type of renewable energy from 2009 to 2018. 

## Grant dollars per PI for each renewable energy 

**Visual on slide 13 of presentation**

```python
grants=grants_df.copy()

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(12,6))
plt.subplots_adjust(hspace=1)
fig.suptitle("Grant Funding Per PI Across Renewable Energy Topics", fontsize=18, y=1)

colors=['skyblue','dodgerblue', 'olivedrab','coral']
RE_types=['solar', 'wind power | wind turbine','biomass | biofuel','geothermal']
for n, RE_type in enumerate(RE_types):
    #iterative subplots
    ax=plt.subplot(2,2,n+1)
    #now filter by each topic's total fiscal year cost throughout the years 
    RE_ppyr1=grants[(grants['Grant_Year'].between(2009,2018)) & (grants['ABSTRACT'].isna()) & (grants['PROJECT_TITLE'].str.contains(RE_type))]
    non_emptyab=grants[grants['ABSTRACT'].notna()]
    RE_ppyr2=non_emptyab[(non_emptyab['Grant_Year'].between(2009,2018)) & (non_emptyab['ABSTRACT'].str.contains(RE_type))]
    combo_df=pd.concat([RE_ppyr1, RE_ppyr2])
    #create a dataframe of grant year with number of unique PIs for each year
    num_PIs=combo_df.groupby('Grant_Year')['CONTACT_PI_PROJECT_LEADER'].nunique().to_frame().reset_index()
    #create a dataframe of grant year with total fiscal year cost for each year
    cost_df=combo_df.groupby('Grant_Year')['FY_TOTAL_COST'].sum().to_frame().reset_index()
    #merge the 2 dataframes on grant year
    merged_PI_cost=cost_df.merge(num_PIs, on='Grant_Year')
    merged_PI_cost['COST_PER_PI']=merged_PI_cost['FY_TOTAL_COST']/merged_PI_cost['CONTACT_PI_PROJECT_LEADER']
    new_df=merged_PI_cost[['Grant_Year', 'COST_PER_PI']]
    new_df['Grant_Year']=new_df['Grant_Year'].astype(int)
    new_df.set_index('Grant_Year', inplace=True) 
    new_df['COST_PER_PI']=new_df['COST_PER_PI']/1e6
    new_df=new_df.rename(columns={'COST_PER_PI':"Funding in USD for "+str(RE_type)+" grants per PI"})
    new_df["Funding in USD for "+str(RE_type)+" grants per PI"].plot(kind='bar', ax=ax, color=colors[n % len(colors)])
    #format the y-axis values & limit
    ax.set_title(str(RE_type))
    ax.set_ylim(0,1)
    ax.set_ylabel('Millions of USD / PI')
    plt.xticks(rotation=65, ha='right')
plt.savefig('FundingperPI_multiplot.jpg', bbox_inches='tight')
plt.show()
```

```python
def group_year_cost(data, date_param):
    years = [2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018]
    data = data.copy()
    data['START_YEAR'] = pd.to_datetime(data['PROJECT_START_DATE']).dt.year
    data['END_YEAR'] = pd.to_datetime(data['PROJECT_END_DATE']).dt.year
    #have an initial dataframe with names and sum of total cost
    result = data.groupby('CONTACT_PI_PROJECT_LEADER').agg({'FY_TOTAL_COST':'sum'}) 
    for year in years:
        name = data.loc[data[date_param] == year].groupby('CONTACT_PI_PROJECT_LEADER').agg({'FY_TOTAL_COST':'sum'}).\
        rename(columns={'FY_TOTAL_COST':year})
        #join the yearly costs with the initial dataframe 
        result = result.join(name)
    return result
```
